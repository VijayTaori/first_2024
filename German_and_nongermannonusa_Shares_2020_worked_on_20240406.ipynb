```python
import pandas as pd

# Sample DataFrame with a column containing sentences
data = {'Sentences': ['This is a sample sentence with numbers 123 and 456.78',
                      'Another example of a sentence with -10 and 3.14',
                      'Yet another example without numbers']}

df = pd.DataFrame(data)

# Splitting sentences into words
word_columns = df['Sentences'].str.split(expand=True)

# Iterate over each column
for col in word_columns.columns:
    # Check if each word in the column can be converted to a number
    try:
        word_columns[col] = pd.to_numeric(word_columns[col])
    except ValueError:
        pass  # Skip if the word cannot be converted to a number

# Rename columns with sequential names like 'a', 'b', 'c', etc.
word_columns.columns = [chr(ord('a') + i) for i in range(len(word_columns.columns))]

# Concatenate original DataFrame with the new word columns
df = pd.concat([df, word_columns], axis=1)

print(df)
```
