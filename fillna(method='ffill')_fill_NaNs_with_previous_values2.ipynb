```python
# Reindex so that all the dates are incorporated
max_date_df_eur = df_usd2eur.index.max()
min_date_df_eur = df_usd2eur.index.min()
df_eur = df_usd2eur.reindex(pd.date_range(min_date_df_eur,max_date_df_eur))
df_eur
```

```python
# Reset index and rename the column generated from resetting to Date
df_eur = df_eur.reset_index().rename(columns={'index':'Date'})
df_eur
```

```python
#Adding Weekday column which includes weekends
df_eur['Date'] = pd.to_datetime(df_eur['Date'],dayfirst=True)
def find_weekday(y):
  day = calendar.day_name[y.weekday()]
  return day
df_eur['Weekday'] = df_eur['Date'].map(find_weekday)
df_eur
```

```python
# Filling NaN with previous values
df_eur.fillna(method='ffill')
```

```python
# Above method has to be inplace=True
# Filling NaN with previous values
df_eur.fillna(method='ffill',inplace=True)
df_eur
```
